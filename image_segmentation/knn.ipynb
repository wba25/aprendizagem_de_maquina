```python
X = data_set.iloc[:,1:].copy()
y = data_set.y.copy()
```

```python
X_test = test_set.iloc[:,1:].copy()
y_test = test_set.y.copy()
```

```python
y.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(X, y)
```

```python
pd.crosstab(neigh.predict(X),y) # s√≥ de teste
```

```python
pd.crosstab(neigh.predict(X_test),y_test)
```

```python
from sklearn.metrics import matthews_corrcoef, classification_report, accuracy_score, roc_curve, roc_auc_score
```

```python
pred_teste = neigh.predict(X_test)
pred_treino = neigh.predict(X)
real_teste = y_test
real_treino = y
print(classification_report(real_treino,pred_treino))
print('MCC treino = {:.3f}, Acuracia treino = {:.3f}'.format(matthews_corrcoef(real_treino,pred_treino),accuracy_score(real_treino,pred_treino)))
print('\n------------------------------\n')
print(classification_report(real_teste,pred_teste))
print('MCC teste = {:.3f}, Acuracia teste = {:.3f}'.format(matthews_corrcoef(real_teste,pred_teste),accuracy_score(real_teste,pred_teste)))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf1 = GaussianNB()
clf1.fit(X, y)
```

```python
clf2 = KNeighborsClassifier(n_neighbors=3)
clf2.fit(X, y)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
eclf = VotingClassifier(estimators=[('knn', clf1), ('gnb', clf2)], voting='hard')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for clf, label in zip([clf1, clf2, eclf], ['k-nearest neighbors','Gaussian Naive Bayes', 'Ensemble']):
    scores = cross_val_score(clf, X, y, cv=5, scoring='accuracy')
    print("Accuracy: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```
